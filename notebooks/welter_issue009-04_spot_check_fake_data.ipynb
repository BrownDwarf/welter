Cool!  Mose of the structure is described by the warm component.  The underlying cool spots are only barely perceptible in the mixture model, even for this extreme absorption band in the cool spectrum.  My guess is that when fitting a single temperature model to this mixture model, most of this coarse structure will be absorbed into the calibration parameters, and most fine structure will be in the noise.  But I think the temperature will get thrown off a little bit.
